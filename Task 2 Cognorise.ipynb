```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\Sh2\Downloads\tested.csv")
df
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python


# Fill missing values in 'Age' with the median and in 'Embarked' with the mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df



```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```

```python


# Drop 'Cabin' since it has too many missing values
df.drop('Cabin', axis=1, inplace=True)
df
```

```python


# Convert categorical columns into numeric ones
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df
```

```python
# Drop irrelevant columns
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python


# Bar plot: Survival rate by Sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.show()
```

```python
# Histogram: Age distribution
sns.histplot(df['Age'], bins=20, kde=True)
plt.show()

```

```python
# Create a new feature 'FamilySize'
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df
```

```python
# Bin the 'Age' feature into categories
df['AgeBin'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, 100], labels=[0, 1, 2, 3])
df.drop('Age', axis=1, inplace=True)
df
```

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4)) 
df['Survived'].value_counts().plot.pie( 
    explode=[0, 0.1], autopct='%1.1f%%', ax=ax[0], shadow=False) 
ax[0].set_title('Survivors (1) and the dead (0)') 
ax[0].set_ylabel('') 
sns.countplot(x='Survived', data=df, ax=ax[1])
ax[1].set_ylabel('Quantity') 
ax[1].set_title('Survivors (1) and the dead (0)') 
plt.show()
```
